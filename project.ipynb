```python
import pandas as pd 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
import matplotlib.pyplot as plt
import numpy as np 
data=pd.read_csv("score_updated.csv")
model=LinearRegression()
X=data[["Hours"]]
y=data[["Scores"]]
model.fit(X,y)
predicted_score=model.predict(X)


mae= mean_absolute_error(y,predicted_score)
mse=mean_squared_error(y,predicted_score)
rmse=np.sqrt(mse)
r2=r2_score(y,predicted_score)

print("MAE(Mean Absolute Eroor):",round(mae,2))
print("MSE(Mean Squared Eroor):",round(mse,2))
print("RMSE(Root Mean Squared Eroor):",round(rmse,2))
print("R^2 SCORE (MODEL ACCURACY):-",round(r2,2))

#HISTOGRAM
plt.figure(figsize=(10,6))
plt.hist(data["Scores"],bins=30,color="skyblue",edgecolor="black")
plt.title("Distribution of FINAL EXAM SCORES")
plt.xlabel("Final Exam Score")
plt.ylabel("No. of Student")
plt.show()


#Scatter Plot and Regression Line
plt.figure(figsize=(10,6))
plt.scatter(X,y,color="blue",label="Actual scores")
plt.plot(X,predicted_score,color="red",label="PREDICTED SCORES(REGRESSION LINE)")
plt.title("Model Prediction VS Actual Score")
plt.xlabel("Study Hours per week")
plt.ylabel("Final Output")
plt.grid(True)
plt.show()

new_hour=float(input("Enter Hours:"))
predicted_new_score=model.predict([[new_hour]])
print(f"Predicted Final Score for {new_hour} is {predicted_new_score}")




```
